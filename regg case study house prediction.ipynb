```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('house_rental_data.csv.txt')
```

```python
data.head()
```

```python
data = data.drop(['Unnamed: 0'], axis = 1)
```

```python
data.head()
```

```python
data.describe(percentiles =[0.01,.5,.1,.25,.75,.9,.95,.99])
```

```python
y = data['Price']
x = data.drop(['Price'],axis =1)
```

```python
def outlier_cap(x):
    x=x.clip(lower=x.quantile(0.01))
    x=x.clip(upper=x.quantile(0.99))
    return(x)

x=x.apply(lambda x : outlier_cap(x))
```

```python
  
from sklearn.feature_selection import VarianceThreshold
varselector= VarianceThreshold(threshold=0)
varselector.fit_transform(x)
# Get columns to keep and create new dataframe with those only
cols = varselector.get_support(indices=True)
x = x.iloc[:,cols]
```

```python
from sklearn.preprocessing import minmax_scale
```

```python
from sklearn.metrics import r2_score, mean_squared_error,mean_absolute_error
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y, test_size = 0.25, random_state = 3)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(xtrain,ytrain)
y_lr_pred = lr.predict(xtest)
print('r2_score for lr',r2_score(ytest,y_lr_pred))
print('Mse for lr',mean_squared_error(ytest,y_lr_pred))
print('Mae for lr',mean_absolute_error(ytest,y_lr_pred))
```

```python
from sklearn.tree import DecisionTreeRegressor
dtree = DecisionTreeRegressor()
dtree.fit(xtrain,ytrain)
y_dtree_pred  = dtree.predict(xtest)
print('r2_score for dtree',r2_score(ytest,y_dtree_pred))
print('Mse for dtree',mean_squared_error(ytest,y_dtree_pred))
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(xtrain,ytrain)
y_rfr_pred  = dtree.predict(xtest)
print('r2_score for rfr',r2_score(ytest,y_rfr_pred))
print('Mse for rfr',mean_squared_error(ytest,y_rfr_pred))
```

```python
from xgboost import XGBRegressor
xgb = XGBRegressor()
xgb.fit(xtrain,ytrain)
y_xgb_pred = xgb.predict(xtest)
print('r2_score for xgb:', r2_score(ytest,y_xgb_pred))
print('Mse for xgb:',mean_squared_error(ytest,y_xgb_pred))
print('Msa for xgb:',mean_absolute_error(ytest,y_xgb_pred))
```

```python
from sklearn.svm import SVR
svr = SVR()
svr.fit(xtrain,ytrain)
y_svr_pred = svr.predict(xtest)
print('r2_score for svr',r2_score(ytest,y_svr_pred))
print('Mse for svr',mean_squared_error(ytest,y_svr_pred))
```
